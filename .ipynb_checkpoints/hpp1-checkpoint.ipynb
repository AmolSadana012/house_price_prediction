```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("housing.csv")
```

```python
data
```

```python
data.info()
```

```python
data.dropna(inplace=True) # To delete the none values. The inplace saves the data object in the data only.
```

```python
data.info()
```

```python
# Split data into Train and Test data
from sklearn.model_selection import train_test_split
x=data.drop(['median_house_value'],axis=1)
y=data['median_house_value']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2)
# 20% of data will be for evaluating and we will not touch this datafor anything.
```

```python
x_test = x_test.reindex(columns=x_train.columns, fill_value=0) 
# To get same no of columns in test data as in train data.
```

```python
# Joining x_train data and y_train data so as to analyse some basic correlation.
train_data=x_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data.select_dtypes(include='number').corr()
# select_dtypes(include='number') removes the string data type columns as .corr() can be performed only on numbers.
```

```python
print(train_data.dtypes)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.select_dtypes(include='number').corr(), annot=True, cmap="YlGnBu")
# annot=annotation
```

# Data Preprocessing

```python
train_data['total_rooms']=np.log(train_data['total_rooms']+1)
train_data['total_bedrooms']=np.log(train_data['total_bedrooms']+1)
train_data['population']=np.log(train_data['population']+1)
train_data['households']=np.log(train_data['households']+1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
# Changing the string to float/integer
train_data.ocean_proximity.value_counts()
```

```python
pd.get_dummies(train_data.ocean_proximity).astype(int)
```

```python
train_data=train_data.join(pd.get_dummies(train_data.ocean_proximity).astype(int)).drop(['ocean_proximity'], axis=1)
```

```python
train_data
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.select_dtypes(include='number').corr(), annot=True, cmap="YlGnBu")
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x="latitude",y="longitude",data=train_data, hue="median_house_value", palette="coolwarm")
# The more red=more expensive house is and the more blue=less expensive.
# As its clear that the houses near cost are more expensive as they are near the ocean.
```

# Feature Engineering

```python
# No. of bedrooms per room?
train_data['bedroom_ratio']=train_data['total_bedrooms']/train_data['total_rooms']
train_data['household_rooms']=train_data['total_rooms']/train_data['households']
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.select_dtypes(include='number').corr(), annot=True, cmap="YlGnBu")
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
x_train, y_train= train_data.drop(['median_house_value'], axis=1),train_data['median_house_value']
reg = LinearRegression()
reg.fit(x_train,y_train)
```
